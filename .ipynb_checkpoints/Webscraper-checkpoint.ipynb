```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

#variables 
first_url = 'https://survivor.fandom.com/wiki/Survivor:_China'
merged_tribe_name = ''
merge_index = False
next_url = 'https://survivor.fandom.com/wiki/Survivor:_Samoa'

def get_soup(url):
    website_url = requests.get(url).text
    soup = BeautifulSoup(website_url,'lxml')
    return soup

def get_table(soup):
    #Get the table that we will use to get castaway info
    My_table = soup.find('table',{'class':'wikitable article-table'})
    return My_table

def get_merge_tribe(soup):
    #Get the name of the merged tribe
    tribe_text = soup.find("div", {"class": "pi-item pi-data pi-item-spacing pi-border-color", "data-source": "tribes"}).get_text(separator=u'|',strip=True)
    merged_tribe_name = tribe_text.split('|')[-1]
    print(tribe_text)
    return merged_tribe_name

def get_next_url(soup):
    #Get the url of the next page to scrape
    next_url = 'https://survivor.fandom.com/wiki/Survivor:_' + soup.find("div", {"class": "pi-item pi-data pi-item-spacing pi-border-color", "data-source": "next"}).find('a', href=True).get_text().replace(' ', '_')
    return next_url
```

```python
def get_next_castaway(array, pointer, prev_pointer):
    clean = array[pointer:pointer+3]
    if(pointer + 6 >= len(array)):
        return len(array) 
    count = pointer + 3
    while count - prev_pointer < 7 or (clean[count-pointer-1] !='' or clean[count-pointer-2] !='' or clean[count-pointer-3]!=''):
        count = count + 1
        clean = array[pointer:count]
    return count
```

```python
def get_cleaned_array(My_table):
    table_text = My_table.get_text()
    table_array = table_text.splitlines()

    #clean up the table text.
    cleaned_array = table_array[14:]
    return(cleaned_array)

def create_dataframe():
    #create the table of castaway info
    column_names = ["Season", "Name", "Age", "Hometown", "Tribe", "Rank", "Day voted out", "Jury place", "Votes against"]
    castaways_by_tribe = pd.DataFrame(columns = column_names)
    return castaways_by_tribe
```

```python
def parse_text(cleaned_array, castaways_by_tribe):
    pointer = 0
    prev_pointer = 0
    merge_index = False
    while len(cleaned_array) > pointer + 3:
        prev_pointer = pointer
        pointer = get_next_castaway(cleaned_array, pointer, prev_pointer)
        castaway_array = filter(None,cleaned_array[prev_pointer:pointer])
        #get the name
        
        if len(castaway_array[0])==0:
            castaway_array = castaway_array[1:]
        
        if castaway_array[0].find(',') != -1:
            name_age = castaway_array[0][0:castaway_array[0].find(',')]
            num_index = re.search(r"\d", name_age)
            name = name_age[0:num_index.start()]
            age = name_age[num_index.start():]
            city_job = castaway_array[0][castaway_array[0].find(',')+2:].split(',')
            city = city_job[0] + city_job[1][0:3] 
            job = city_job[1][3:] 
            tribe = []
            index = 1
            #get a list of tribes that the castaways have been in
            
            while ((not castaway_array[index][1].strip().isdigit()) or (not castaway_array[index][0].strip().isdigit()) )and (not castaway_array[index] == ' SecondRunner-Up') and (not 'Evac' in castaway_array[index]) and (not 'Quit' in castaway_array[index]) and (not castaway_array[index] == ' Runner-Up')  and (not castaway_array[index] == ' Sole Survivor'):
                print('test0' , castaway_array[index])
                if len(castaway_array[index]) > 3:
                    #check if it is the merged tribe name
                    print('strip', merged_tribe_name.strip())
                    if castaway_array[index].strip() == merged_tribe_name.strip():
                        merge_index = True
                    tribe.append(castaway_array[index])
                index = index + 1
                print('ind', index)
                print(castaway_array)
                print('test' , len(castaway_array[index]))
            #if the merge tribe index is true, add the merge tribe name to the list of tribes
            if merge_index == True:
                tribe.append(merged_tribe_name)
            #get the place of the survivor
            if (not castaway_array[index] == ' SecondRunner-Up') and (not castaway_array[index] == ' Runner-Up')  and (not castaway_array[index] == ' Sole Survivor'):
                print(name)
                print('nnnn' ,castaway_array[index])
                if 'Quit' in castaway_array[index]:
                    placing = 'Quit'
                    voted_out = castaway_array[index].find('Quit')
                    num = 4
                elif 'Evac' in castaway_array[index]:
                    print('evvvac')
                    placing = 'Evacuated'
                    voted_out = castaway_array[index].find('Evacuated')
                    num = 9
                else :
                    voted_out = castaway_array[index].find('Voted')
                    num = 9
                day = castaway_array[index].find('Day')
                print(voted_out)
                placing = castaway_array[index][0:voted_out+num]
                #get the jury placement
                if 'Jury' in castaway_array[index]:
                    jury_member = castaway_array[index][voted_out+num:day]
                else:
                    jury_member = 'Voted out pre-jury'
                #get the days in the game
                days_in_game = castaway_array[index][day:]
            else:
                placing = castaway_array[index]
                jury_member = 'Made it to final council'
                days_in_game = 'Day 39'
            #get the votes against
            print('ben', castaway_array[index])
            votes_against = castaway_array[index+1]
            #add info the the array
            for tribe_name in tribe:
                row = {'Season': first_url.replace('https://survivor.fandom.com/wiki/Survivor:_', ''), 'Name':name, 'Age':age, 'Hometown':city, 'Tribe':tribe_name, 'Rank':placing, 'Day voted out':days_in_game, 'Jury place':jury_member, 'Votes against': votes_against}
                castaways_by_tribe = castaways_by_tribe.append(row, ignore_index=True) 
    return castaways_by_tribe
```

```python
soup = get_soup(next_url)
table = get_table(soup)
merged_tribe_name = get_merge_tribe(soup)
first_url = next_url
next_url = get_next_url(soup)
#dataframe and cleaned array
cleaned_array = get_cleaned_array(table)
castaways_by_tribe = create_dataframe()
print(cleaned_array)
#run parse text
output_table = parse_text(cleaned_array, castaways_by_tribe)
#Quit
#Evac
#[u'Chicken Morris47, Marion, VAChicken Farmer', u' Zhan Hu', u'', u'', u' 1st Voted OutDay 3', u' 5', u'', u'',
#[u'Jon Dalton33, Los Angeles, CAPearl Islands', u' Malakal', u'', u'', u' 1st Voted OutDay 3', u' 9', u'', u'', 
#[u'', u'Michelle Chase24, Los Angeles, CAMusic Producer', u' Fang', u'', u'', u'', u' 1st Voted OutDay 3', u' 8', 
```
